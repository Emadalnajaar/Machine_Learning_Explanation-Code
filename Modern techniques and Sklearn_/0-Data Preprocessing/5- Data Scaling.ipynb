## 1- StandardScaler (Standarization)
* work as : (value - avarage)/StandardVision

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
data = [[2320, 4420], [4240, 314], [21351, 5121], [2151, 13]]
data = SC.fit_transform(data)
data
```

## 2- MinMaxScaler (Normalization)
* work as : (value - avarage)/(maxVal-minVal)

```python
from sklearn.preprocessing import MinMaxScaler
MinMax_SC = MinMaxScaler(feature_range=(-1, 1))
# default feature_range=(0, 1)
data = [[2320, 4420], [4240, 314], [21351, 5121], [2151, 13]]
data = MinMax_SC.fit_transform(data)
data
```

# 3- Normalizer
* for each `ROW`

```python
Image(filename='norm.png')
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer(norm= 'l1')
data = [[2320, 4420], [4240, 314], [21351, 5121], [2151, 13]]
data = norm.fit_transform(data)
data
```

# 4- MaxAbsScaler
* for each `COL` ==> making max value is the absValue

```python
from sklearn.preprocessing import MaxAbsScaler
MAS = MaxAbsScaler()
data = [[ 1., 10., 2.],
     [ 2., 0., 0.],
     [ 5., 1., -1.]]
data = MAS.fit_transform(data)
data
```

## 5- FunctionTransformer
* take a function for scaling

```python
from sklearn.preprocessing import FunctionTransformer
FT = FunctionTransformer(func= lambda x: x**0.1 , validate= True)

data = [[4, 1, 2, 2], [1, 3, 9, 3], [5, 7, 5, 1]]
data = FT.fit_transform(data)

data
```

# 6- Binarizer
* take a value called threshold , if any value > threshold = 1 else = 0

```python
from sklearn.preprocessing import Binarizer
BZ = Binarizer(threshold= 5.0)

data = [[4, 1, 2, 2], [1, 3, 9, 3], [5, 7, 5, 1]]

data = BZ.fit_transform(data)

data
```

## 7- PolynomialFeatures
* for adding a new feature
* if we have 2 feature and degree = 2 ==> 1,a,b,a^2,ab,b^2

```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

# PF_Scaler = PolynomialFeatures(degree=2, include_bias= True, interaction_only= False)
# include_bias = True ==> FOR X0 = 1
# interaction_only = True ==> show a*b and remove a^2 , b^2

data = np.arange(6).reshape(3,2)
print(data,'\n')

poly = PolynomialFeatures(degree=2, include_bias=True)
data = poly.fit_transform(data)
print(data,'\n')
# 1row, 2row, 3row, 4row   , 5row     , 6row 
# 1   , 2row, 3row, 1row^2 , 1row*2row, 2row^2 ,
```

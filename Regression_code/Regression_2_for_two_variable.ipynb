```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

path = '/Users/fox/Desktop/ML_Art/Regression_code/data2.txt'
data = pd.read_csv(path, header = None, names=['Size', 'Bedrooms' , 'Price'])
```

## Rescaling data
because the range of data is so far for each other, so we need to rescale it by make all in a range between [-1,1]

```python
data = (data - data.mean()) / data.std() # the formula
data.head(5)
```

```python
data.describe()
```

## add one column

```python
data.insert(0, 'Ones' , 1)
data.head(5)
```

##  separate X (training data) from y (target variable)

```python
cols = data.shape[1]
X = data.iloc[:,0:cols-1]
y = data.iloc[:,cols-1:cols]
```

## convert to matrices and initialize theta

```python
X = np.matrix(X.values)
y = np.matrix(y.values)
theta = np.matrix(np.array([0,0,0]))
```

## Cost function

```python
def computeCost(X, y, theta):
    inner = np.power(((X * theta.T) - y), 2)
    return np.sum(inner) / (2 * len(X))
```

## Gradient Descent

```python
 def gradientDescent(X, y, theta, alpha, iters):
    temp = np.matrix(np.zeros(theta.shape))
    parameters = int(theta.ravel().shape[1])
    cost = np.zeros(iters)
    
    for i in range(iters):
        error = (X * theta.T) - y
        
        for j in range(parameters):
            term = np.multiply(error, X[:,j])
            temp[0,j] = theta[0,j] - ((alpha / len(X)) * np.sum(term))
            
        theta = temp
        cost[i] = computeCost(X, y, theta)
        
    return theta, cost
```

## initialize variables for learning rate and iterations

```python
alpha = 0.1
iters = 100

# perform linear regression on the data set
g, cost = gradientDescent(X, y, theta, alpha, iters)

# get the cost (error) of the model
thiscost = computeCost(X, y, g)


print('g2 = ' , g)
print('---------------------------')
print('cost2  = ' , cost[0:20] )
print('---------------------------')
print('computeCost = ' , thiscost)
print('**************************************')
```

## get best fit line for Size vs. Price

```python
x = np.linspace(data.Size.min(), data.Size.max(), 100)
print('x \n',x[:10])
print('---------------------------')
print('g \n',g[:10])
print('---------------------------')

f = g[0, 0] + (g[0, 1] * x)
print('f \n',f[:10])
```

## draw the line for `Size vs. Price`

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, f, 'r', label='Prediction')
ax.scatter(data.Size, data.Price, label='Training Data')
ax.legend(loc=2)
ax.set_xlabel('Size')
ax.set_ylabel('Price')
ax.set_title('Size vs. Price')
```

## get best fit line for Bedrooms vs. Price

```python
x = np.linspace(data.Bedrooms.min(), data.Bedrooms.max(), 100)
print('x \n',x[:10])
print('---------------------------')
print('g \n',g[:10])
print('---------------------------')

f = g[0, 0] + (g[0, 1] * x)
print('f \n',f[:10])
```

## draw the line  for `Bedrooms vs. Price`

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, f, 'r', label='Prediction')
ax.scatter(data.Bedrooms, data.Price, label='Traning Data')
ax.legend(loc=2)
ax.set_xlabel('Bedrooms')
ax.set_ylabel('Price')
ax.set_title('Size vs. Price')
```

## draw error graph

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(np.arange(iters), cost, 'r')
ax.set_xlabel('Iterations')
ax.set_ylabel('Cost')
ax.set_title('Error vs. Training Epoch')
```

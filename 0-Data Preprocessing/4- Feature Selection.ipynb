## 1- Select Percentile
* that select the important feature that correlated with the result for the  percentage by `chi2` , `f_classif` 

```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectPercentile ,chi2, f_classif

breast_cancer = load_breast_cancer()
X= breast_cancer.data
y= breast_cancer.target

print('X shape before is {}\n'.format(X.shape))
 
feature_selection = SelectPercentile(score_func= chi2, percentile= 30)
X_new = feature_selection.fit_transform(X,y)
print('X shape after is {}\n'.format(X_new.shape))

# to show which feature it taken
print('Selected Features are:\n{}\n : '.format(feature_selection.get_support()))
print('Names of  Features Slected are:\n{}\n : '.format(breast_cancer.feature_names))

print('-------------------------------------')
print('Names of  Features Slected are:\n{}\n : '.format(breast_cancer.feature_names[feature_selection.get_support()]))


```

## 2- Univariate Selection

```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import GenericUnivariateSelect ,chi2

breast_cancer = load_breast_cancer()
X= breast_cancer.data
y= breast_cancer.target

print('X shape before is {}\n'.format(X.shape))
 
feature_selection = GenericUnivariateSelect(chi2, 'k_best', param=10)
X_new = feature_selection.fit_transform(X,y)
print('X shape after is {}\n'.format(X_new.shape))

# to show which feature it taken
print('Selected Features are:\n{}\n : '.format(feature_selection.get_support()))
print('Names of  Features Slected are:\n{}\n : '.format(breast_cancer.feature_names))

print('-------------------------------------')
print('Names of  Features Slected are:\n{}\n : '.format(breast_cancer.feature_names[feature_selection.get_support()]))


```

## 3- K-best Selection

```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectKBest, chi2

breast_cancer = load_breast_cancer()
X= breast_cancer.data
y= breast_cancer.target

print('X shape before is {}\n'.format(X.shape))
 
feature_selection = SelectKBest(chi2, k = 3)
X_new = feature_selection.fit_transform(X,y)
print('X shape after is {}\n'.format(X_new.shape))

# to show which feature it taken
print('Selected Features are:\n{}\n : '.format(feature_selection.get_support()))
print('Names of  Features Slected are:\n{}\n : '.format(breast_cancer.feature_names))

print('-------------------------------------')
print('Names of  Features Slected are:\n{}\n : '.format(breast_cancer.feature_names[feature_selection.get_support()]))


 
```

## 4- Select From Model

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
X = data.data
y = data.target
print('X shape before is {}\n'.format(X.shape))
 
LR = LinearRegression()    
feature_selection = SelectFromModel(estimator= LR)
# in SelectFromModel take any model and estimate number of feature on it 

X_new = feature_selection.fit_transform(X,y)
print('X shape after is {}\n'.format(X_new.shape))

# to show which feature it taken
print('Selected Features are:\n{}\n : '.format(feature_selection.get_support()))
print('Names of  Features Slected are:\n{}\n : '.format(breast_cancer.feature_names))

print('-------------------------------------')
print('Names of  Features Slected are:\n{}\n : '.format(breast_cancer.feature_names[feature_selection.get_support()]))
```

## 5- Recursive Feature Elimination Selection 'RFE'
* the best one for making a ranking of the best

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
X = data.data
y = data.target
print('X shape before is {}\n'.format(X.shape))
 
LR = LinearRegression()    
feature_selection = RFE(estimator= LR)
# in SelectFromModel take any model and estimate number of feature on it 

X_new = feature_selection.fit_transform(X,y)
print('X shape after is {}\n'.format(X_new.shape))

# to show which feature it taken
print('Selected Features are:\n{}\n : '.format(feature_selection.get_support()))
print('Names of  Features Slected are:\n{}\n : '.format(breast_cancer.feature_names))

print('-------------------------------------')
print('Names of  Features Slected are:\n{}\n : '.format(breast_cancer.feature_names[feature_selection.get_support()]))

print('Names of  Features Slected are:\n{}\n : '.format(feature_selection.ranking_))
```
